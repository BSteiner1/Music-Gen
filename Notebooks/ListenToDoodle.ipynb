```python
# Create an instance of the PhraseToMIDI class
midi_class = PhraseToMIDI(sample_phrase, tempo=200)

# Call the write_phrase_to_MIDI method to generate MIDI data
sample_midi_data = midi_class.write_phrase_to_MIDI(sample_phrase)
```

```python
def display_audio(pm: pretty_midi.PrettyMIDI, seconds=30):
  waveform = pm.fluidsynth(fs=20000)
  # Take a sample of the generated waveform to mitigate kernel resets
  waveform_short = waveform[:seconds*20000]
  return display.Audio(waveform_short, rate=20000)
```

```python
display_audio(sample_midi_data)
```
